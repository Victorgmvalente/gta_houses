-Define df_text que vai ter os valores contidos na coluna 'text' do dataframe
-para cada linha 'i' no comprimento do dataframe df2: <br>
    -define js com os valores do df2 coluna text/linha i e retorna os valores contidos em 'response', 'content', e 'text' <br>
    -define json_object como um json serialiazed de js <br>
    -insere json_object na coluna text <br>

```python
df_text = pd.DataFrame(columns = {'text':''})
for i in range(0,len(df2)):
    js = df2['text'][i].get('response',{}).get('content',{}).get('text')
    json_object = json.dumps(js)
    #item = str(i)
    df_text = df_text.append({'text':json_object}, ignore_index = True)
```
